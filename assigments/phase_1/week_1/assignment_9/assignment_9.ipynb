```python
from datasets import load_dataset
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB   
from sklearn.metrics import classification_report, accuracy_score

print("Load du lieu")
dataset = load_dataset("clapAI/MultiLingualSentiment")

filter_dataset = dataset.filter(lambda x: x['language'] in ['vi', 'en'])
X_train = list(filter_dataset['train']['text'])
y_train = list(filter_dataset['train']['label'])

X_test = list(filter_dataset['test']['text'])
y_test = list(filter_dataset['test']['label'])

print(f"Xem thu du lieu")
print(f"Van ban: \"{X_train[0]}\"")
print(f"Nhan: {y_train[0]}")

print(f"Xem thu du lieu test")
print(f"Van ban: \"{X_test[0]}\"")
print(f"Nhan: {y_test[0]}")

print("\nĐang chuyển đổi văn bản sang ma trận thưa BoW...")
bow_vectorizer = CountVectorizer(
    ngram_range=(1, 2), 
    max_features=100000,
    stop_words=['english', 'vietnamese']
)
X_train_bow = bow_vectorizer.fit_transform(X_train)
X_test_bow = bow_vectorizer.transform(X_test)

bow_vectorizer.get_feature_names_out()

print("\nĐang chuyển đổi văn bản sang ma trận thưa TFIDF...")
tfidf_vectorizer = TfidfVectorizer(
    max_features=100000,
    stop_words=['english', 'vietnamese'],
    ngram_range=(1, 2)
)
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

tfidf_vectorizer.get_feature_names_out()
```

```python
print("Logistic Regression")
print("BOW")
bow_model = LogisticRegression(
    C=0.8,
    max_iter=100,           
    solver='saga',
    class_weight='balanced',
    warm_start=True,     
    verbose=1
)

bow_model.fit(X_train_bow, y_train)

print("\nHuan luyen xong")

y_test_pred = bow_model.predict(X_test_bow)

print(f"Accuracy: {accuracy_score(y_test, y_test_pred):.4f}\n")
print("Report:")
print(classification_report(y_test, y_test_pred))

print("TFIDF")
tfidf_model = LogisticRegression(
    C=0.8,
    max_iter=100,           
    solver='saga',
    class_weight='balanced',
    warm_start=True,   
    verbose=1
)

tfidf_model.fit(X_train_tfidf, y_train)

print("Huan luyen xong")

y_pred = tfidf_model.predict(X_test_tfidf)

print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}\n")
print("Report:")
print(classification_report(y_test, y_pred))
```

```python
print("Naive Bayes")
print("BOW")
bow_naivebayes = MultinomialNB(alpha=1)

bow_naivebayes.fit(X_train_bow, y_train)

print("Huan luyen xong")
y_test_pred_bow = bow_naivebayes.predict(X_test_bow)

print(f"Accuracy: {accuracy_score(y_test, y_test_pred_bow):.4f}\n")
print("Report:")
print(classification_report(y_test, y_test_pred_bow))

print("TFIDF")
tfidf_naivebayes = MultinomialNB(alpha=1)

tfidf_naivebayes.fit(X_train_bow, y_train)

print("Huan luyen xong")
y_test_pred_tfidf = tfidf_naivebayes.predict(X_test_bow)

print(f"Accuracy: {accuracy_score(y_test, y_test_pred_tfidf):.4f}\n")
print("Report:")
print(classification_report(y_test, y_test_pred_tfidf))
```
